```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the graph with 8 vertices
G = nx.Graph()

# Add vertices
vertices = range(1, 9)
G.add_nodes_from(vertices)

# Add edges with weights
edges = [
    (1, 2, 3), (1, 3, 2), (1, 4, 1),
    (2, 5, 4), (3, 5, 5), (3, 6, 6),
    (4, 6, 2), (4, 7, 3), (5, 8, 1),
    (6, 8, 4), (7, 8, 2)
]
G.add_weighted_edges_from(edges)

# Initial matching (just for example)
initial_matching = [(1, 2), (3, 5), (4, 6)]

# Augmenting path (example)
augmenting_path = [(1, 3), (3, 5), (5, 8)]

# Increasing path (example)
increasing_path = [(3, 6), (6, 8), (8, 7)]

# Function to draw graph with highlighted edges
def draw_graph(G, highlighted_edges=None, title="Graph"):
    pos = nx.spring_layout(G)
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=16)
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    if highlighted_edges:
        nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='r', width=2)

    plt.title(title)
    plt.show()

# Draw initial graph
draw_graph(G, title="Initial Graph")

# Draw graph with initial matching
draw_graph(G, highlighted_edges=initial_matching, title="Graph with Initial Matching")

# Draw graph after augmenting path
draw_graph(G, highlighted_edges=augmenting_path, title="Graph with Augmenting Path")

# Draw graph after increasing path
draw_graph(G, highlighted_edges=increasing_path, title="Graph with Increasing Path")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with 9 vertices
G = nx.Graph()

# Add vertices with weights
vertex_weights = {
    1: 10, 2: 8, 3: 6,
    4: 7, 5: 9, 6: 5,
    7: 4, 8: 3, 9: 2
}

# Add edges with weights
edges = [
    (1, 2, 5), (1, 3, 3), (2, 4, 7), (3, 5, 8),
    (4, 6, 4), (5, 7, 6), (6, 8, 2), (7, 9, 1),
    (8, 9, 3), (2, 5, 4), (3, 6, 6), (1, 4, 9)
]

# Add nodes to the graph
for v, weight in vertex_weights.items():
    G.add_node(v, weight=weight)

# Add edges to the graph
for u, v, weight in edges:
    G.add_edge(u, v, weight=weight)

# Draw the graph before matching
pos = nx.spring_layout(G)
labels = {v: f"{v}({w})" for v, w in vertex_weights.items()}
edge_labels = {(u, v): d["weight"] for u, v, d in G.edges(data=True)}

plt.figure(figsize=(18, 6))

# Plot 1: Graph before matching
plt.subplot(131)
nx.draw(G, pos, with_labels=True, labels=labels, node_size=700, node_color="lightblue")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Before Matching")

# Create an initial matching (arbitrarily chosen for illustration)
initial_matching = [(1, 2), (3, 5), (4, 6)]

# Draw the graph with the initial matching
matching_edges = [(u, v) for u, v in initial_matching]
non_matching_edges = [e for e in G.edges if e not in matching_edges]

# Plot 2: Graph after augmenting path
plt.subplot(132)
nx.draw(G, pos, with_labels=True, labels=labels, node_size=700, node_color="lightblue")
nx.draw_networkx_edges(G, pos, edgelist=matching_edges, edge_color="red", width=2)
nx.draw_networkx_edges(G, pos, edgelist=non_matching_edges, edge_color="gray", style="dashed")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("After Augmenting Path")

# Create a new matching after increasing path (arbitrarily chosen for illustration)
new_matching = [(1, 4), (3, 6), (5, 7)]

# Draw the graph with the new matching
matching_edges = [(u, v) for u, v in new_matching]
non_matching_edges = [e for e in G.edges if e not in matching_edges]

# Plot 3: Graph after increasing path
plt.subplot(133)
nx.draw(G, pos, with_labels=True, labels=labels, node_size=700, node_color="lightblue")
nx.draw_networkx_edges(G, pos, edgelist=matching_edges, edge_color="green", width=2)
nx.draw_networkx_edges(G, pos, edgelist=non_matching_edges, edge_color="gray", style="dashed")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("After Increasing Path")

plt.tight_layout()
plt.show()
```
